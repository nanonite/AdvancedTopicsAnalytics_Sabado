Analyze the writing patterns of each student

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
from textstat import flesch_reading_ease

# Descargar palabras vacías en español
nltk.download('stopwords')
stop_words_spanish = set(stopwords.words('spanish'))

# Supongamos que tienes un DataFrame llamado df con cinco columnas de texto: 'columna1', 'columna2', 'columna3', 'columna4', 'columna5'

# Tokenización y análisis de frecuencia
def tokenize_and_frequency(text):
    if isinstance(text, str):
        tokens = word_tokenize(text.lower())  # Convertir a minúsculas y tokenizar
        tokens = [word for word in tokens if word.isalnum()]  # Eliminar puntuación
        filtered_tokens = [word for word in tokens if word not in stop_words_spanish]  # Eliminar palabras vacías en español
        return filtered_tokens
    else:
        return []

# Análisis de sentimientos
def sentiment_analysis(text):
    sia = SentimentIntensityAnalyzer()
    sentiment_score = sia.polarity_scores(text)
    return sentiment_score

def readability_analysis(text):
    if isinstance(text, str):
        readability_score = flesch_reading_ease(text)
        return readability_score
    else:
        return 0  # Return a default readability score if the input is not a string

    return readability_score

# Aplicar funciones a cada columna del DataFrame
for column in data.columns:
    # Tokenización y análisis de frecuencia
    data[f'{column}_tokens'] = data[column].apply(tokenize_and_frequency)

    # Análisis de sentimientos
    #data[f'{column}_sentiment'] = data[column].apply(sentiment_analysis)

    # Análisis de legibilidad
    data[f'{column}_readability'] = data[column].apply(readability_analysis)

# Ahora el DataFrame contiene columnas adicionales con tokens, puntuaciones de sentimiento y puntuaciones de legibilidad para cada columna de texto

# Puedes acceder a estos datos adicionales utilizando las nuevas columnas agregadas al DataFrame
print(data.head())
```



El puntaje de legibilidad de Flesch varía en una escala de 0 a 100, donde puntajes más altos indican un texto más fácil de leer. Aquí tienes una interpretación general de los puntajes:

    90-100: Muy fácil de leer. Adecuado para estudiantes de primaria.
    80-90: Fácil de leer. Adecuado para estudiantes de secundaria.
    70-80: Moderadamente fácil de leer. Adecuado para lectores de nivel universitario.
    60-70: Adecuado para lectores adultos.
    50-60: Adecuado para lectores con un nivel educativo de secundaria.
    30-50: Adecuado para lectores con un nivel educativo universitario.
    0-30: Difícil de leer. Adecuado para lectores altamente educados.

```python
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# Ejemplo de DataFrame con columnas de texto

c=['E1 - Examples of Time Series Analysis', 'E4 - Panel Data',
       'E6 - Decison Trees', 'E8 - Ensemble Trees Overview', 'E10 - REST-API',
       'E11 - Natural Language Processing Overview']



df = data[c]
df = df.astype(str)


# Entrenar un modelo Word2Vec
sentences = [text.split() for text in df.values.flatten()]
model = Word2Vec(sentences, min_count=1, vector_size=100)

# Calcular los vectores de representación de cada columna
column_vectors = [model.wv[text.split()].mean(axis=0) for text in df['E1 - Examples of Time Series Analysis']]

# Calcular la similitud coseno entre las columnas
similarities = cosine_similarity(column_vectors)

# Convertir la matriz de similitud en un DataFrame

similarities
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

import pandas as pd
import unidecode
import spacy

# Supongamos que tu DataFrame es df con columnas 'A', 'B', y 'C'

# Preprocesamiento de texto
stop_words = nltk.download('stopwords')

nltk.download('stopwords')
stop_words_spanish = set(stopwords.words('spanish'))

# Supongamos que tienes un DataFrame llamado df con cinco columnas de texto: 'columna1', 'columna2', 'columna3', 'columna4', 'columna5'

# Tokenización y análisis de frecuencia
def tokenize_and_frequency(text):
    if isinstance(text, str):
        tokens = word_tokenize(text.lower())  # Convertir a minúsculas y tokenizar
        tokens = [word for word in tokens if word.isalnum() and not word.isdigit()]  # Eliminar puntuación y números
        filtered_tokens = [word for word in tokens if word not in stop_words_spanish]  # Eliminar palabras vacías en español
        return filtered_tokens
    else:
        return []
# Supongamos que tu DataFrame es df y contiene la columna 'combined_text'

# Función para eliminar tildes y caracteres especiales
def eliminar_tildes(texto):
    texto_sin_tildes = unidecode.unidecode(texto)
    return texto_sin_tildes

# Función para lematizar el texto en español
def lematizar_texto(texto):
    nlp = spacy.load("es_core_news_sm")
    doc = nlp(texto)
    texto_lematizado = ' '.join([token.lemma_ for token in doc])
    return texto_lematizado




data['texto_limpio'] = data['combined_text'].apply(eliminar_tildes)
data['texto_limpio'] = data['texto_limpio'].apply(lematizar_texto)


data['clean_text'] = data['texto_limpio'].apply(tokenize_and_frequency)



# Aplicar limpieza de tildes y lematización al texto combinado





# División de datos
X_train, X_test, y_train, y_test = train_test_split(data['clean_text'], data['Genero'], test_size=0.3, random_state=42)

```

```python
X_train_flattened = [' '.join(words) for words in X_train]
X_test_flattened = [' '.join(words) for words in X_test]

# Vectorización de texto
vectorizer = TfidfVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train_flattened)
X_test_vectors = vectorizer.transform(X_test_flattened)
```

```python
df_test_vectors = pd.DataFrame(X_test_vectors.toarray(), columns=vectorizer.get_feature_names_out())

# Visualize the DataFrame
print(df_test_vectors.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Construcción del modelo clasificador con ajuste de hiperparámetros
classifier = RandomForestClassifier(n_estimators=20, max_depth=10, random_state=42)

# Entrenamiento del modelo
classifier.fit(X_train_vectors, y_train)

# Predicción del modelo en el conjunto de prueba
predictions = classifier.predict(X_test_vectors)

# Evaluación del modelo
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```
